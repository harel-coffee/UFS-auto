```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.style.use('seaborn-dark')

sample_number_parameter = ['1200','1800','2400','3000']

fig = plt.figure(figsize = (10,6))

plt.grid(True,linestyle='--',linewidth = 2.5,zorder=-1,axis="y")
plt.xticks(np.arange(len(sample_number_parameter)), sample_number_parameter,fontsize = 28,rotation=0)

plt.xlabel('Number of training samples', fontsize = 28)

test_error=np.array([0.023769885004783164,0.023039705237563775,0.022829714405293366,0.022675320372289542])

plt.plot(np.arange(len(sample_number_parameter)), test_error, marker='o', mec='blue',c='blue',mfc='w',ms=12)


plt.yticks(fontsize=28)
plt.ylabel('Test error',fontsize=28)
plt.yticks(np.arange(0.0225, 0.0242,0.0004),['2.25E-2','2.29E-2','2.33E-2','2.37E-2','2.41E-2'],fontsize=28)
plt.ylim([0.0225, 0.0242])

plt.show()
```
