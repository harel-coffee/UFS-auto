```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
seed=0
test_num_used=288

np.random.seed(seed)
test_num=288
index_=np.arange(test_num)
np.random.shuffle(index_)
```

```python
index=index_[0:test_num_used]
```

```python
index
```

```python
samples_=np.arange(100,1100,200)
```

```python
samples_
```

```python
results_collection=[]
for i in samples_:
    sample_used_1=i
    resutls_get_1_=np.array(pd.read_csv("./log/results_"+str(sample_used_1)+".csv",header=None))
    resutls_get_1=resutls_get_1_[:,0][index]
    
    sample_used_2=i-1
    resutls_get_2_=np.array(pd.read_csv("./log/results_"+str(sample_used_2)+".csv",header=None))
    resutls_get_2=resutls_get_2_[:,0][index]
    print(i," vs ", (i-1))
    results_collection_i=np.max(np.abs(resutls_get_1-resutls_get_2))
    print(results_collection_i)
    results_collection.append(results_collection_i)
```

```python
samples_
```

```python
results_collection
```

```python
ms=['*','D','v','s','p','X','o','>','<','^','1','3','4','8','9','H','d','+','2','|','_']
cs=['red','green','purple','blue','orange','aqua','darkcyan','olive','darkcyan','silver','pink']

plt.figure(figsize=(10,6))

plt.grid(True,linestyle='--',linewidth = 1.5,zorder=-1)

plt.xticks(samples_, fontsize = 35)
plt.xlabel('#Samples', fontsize = 35)
plt.yticks(np.arange(0.0,0.07,0.02), fontsize = 35)
plt.ylabel(r'$\beta$', fontsize = 35)
plt.ylim(0.0,0.07)

plt.plot(samples_, results_collection, marker=ms[0],linewidth = 3.5,c=cs[0],ms=40)

#plt.legend(labels=Method_title,fontsize=35, loc='upper center', bbox_to_anchor=(1.3,1.1),ncol=1,handletextpad=0.1,columnspacing=1.2, fancybox=True,framealpha=0.1,shadow=False)
        
plt.show()
```
